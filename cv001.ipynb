```python
# let's record a video 
# cv.VideoCapture(0) 0 => means you are going to capture from the device
# cv.VideoCapture('path//') => using an existed video
x, y = 640, 480
cap = cv2.VideoCapture(0)
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('sr_rec001.avi', fourcc, 20.0, (x, y))
print(cap.isOpened())
while cap.isOpened():
    ret, frame = cap.read()
    #if ret is True then their is a frame to be saved
    #else their is no frames
    if ret == True:
        print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        out.write(frame)

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv2.imshow('frame', gray)

        if cv2.waitKey(1) == ord('q'):
            break
    else :
        break
   
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
greenPizza = cv2.imread('greenPizza.jpeg', 1)
green_copy = np.copy(greenPizza)
cv2.imshow('color',  green_copy)
# so note ðŸ’¡ 
# openCV reads colors images as BGR 
# pyplot reads ~ ~ as           RGB

green_copy = cv2.cvtColor(green_copy, cv2.COLOR_BGR2RGB)
plt.imshow(green_copy);

if cv2.waitKey(0) == ord('x'):
    cv2.destroyAllWindows()



print(type(greenPizza), greenPizza.shape, greenPizza.size)
```

# Define ColorðŸŽ¨ threshold


```python
# define our colors boundaries 
lower_green = np.array([0, 100, 0])
upper_green = np.array([70, 255, 50])
```

# Create a Mask ðŸŽ­ 

```python
# define the mask area
mask = cv2.inRange(green_copy, lower_green, upper_green) # asks wether a color falls in the range or not
# if it's in the range it would be displayed else blocked out
# visual the mask 
plt.imshow(mask, cmap='gray');
```

```python
# mask the image to let anything show
masked_image = np.copy(green_copy)
masked_image [mask != 0] = [0, 0, 0]
plt.imshow(masked_image);
```
